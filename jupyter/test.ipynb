```python
"""
问题1: 求每个team的季度平均值

1. 需要对team进行分组
2. 利用mean对数据进行求平均值
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv("team.csv")
df.columns

# df.groupby("team").sum()
teamMean = df.groupby("team").mean(numeric_only=True) # numeric_only=True 为了不报错

print(teamMean)
teamMeanT =  teamMean.T
print(teamMeanT)


# plt.xticks(ticks=["Q1","Q2","Q3","Q4"],labels=["Q1","Q2","Q3","Q4"])



# plt.plot(teamMeanT["A"],label="A")
# plt.plot(teamMeanT["B"],label="B")
# plt.plot(teamMeanT["C"],label="C")
# plt.plot(teamMeanT["D"],label="D")
# plt.legend()

team = ["A","B","C","D","E"]

for i in range (0,5):
    plt.plot(teamMeanT[team[i]],label=team[i],marker="o")


# 更改y坐标轴的位置
ax=plt.gca()
ax.yaxis.set_major_locator(plt.MultipleLocator(5))

plt.legend()
plt.grid(axis='y',linestyle = '--')
plt.show()

```
